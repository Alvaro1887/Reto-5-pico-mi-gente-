```python
import math

def medidas_figuras(r, alto, ancho):
    def calculo_area():
        area_circunferencia = 2 * (math.pi * (r ** 2))
        area_rectangulo = alto * ancho 

        return area_circunferencia + area_rectangulo
    return calculo_area()

def calcular_perimetro(r, alto, ancho):
    perimetro_circunferencia = (2 * r) * math.pi
    perimetro_rectangulo = (2 * alto) + (2 * ancho)
    return perimetro_circunferencia + perimetro_rectangulo

def main():
    r = float(input("Ingresa el radio del circulo mi querido amigo (r): "))
    alto = float(input("Ingresa la altura del rectangulo (alto): "))
    ancho = float(input("Ingrese el ancho del rectangulo (ancho): "))

    area = medidas_figuras(r, alto, ancho) 
    perimetro = calcular_perimetro(r, alto, ancho)

    print(f"Area total: {area:.2f} cm²")
    print(f"Perimetro total: {perimetro:.2f}")

if __name__ == "__main__":
    main()
```

## Punto 3
Diseñe una función que calcule la cantidad de carne de aves en kilos si se tienen N gallinas, M gallos y K pollitos cada uno pesando 6 kilos, 7 kilos y 1 kilo respectivamente.

```python

import math

def animales_sabrosos_xd(N, M, K):
    peso_gallina= 6
    peso_gallo= 7
    peso_pollito= 1

    total_carne = (N * peso_gallina) + (M * peso_gallo) + (K * peso_pollito)
    return total_carne

def main ():
    N = int(input("Ingresa la cantidad de gallinas :) (N): "))
    M = int(input("Ingresa los gallos mi querido amigo (M): "))
    K = int(input("Ingresa los pollos mi querido amigo (K): "))

    total_carne = animales_sabrosos_xd(N, M, K)

    print(f"La cantidad total de carne de aves es: {total_carne} kilos")

if __name__ == "__main__":
    main()
     
```

## Punto 6
Dado el centro y el radio de un círculo, determinar si un punto de R2 pertenece o no al interior del círculo

```python
import math

def esta_en_el_circulo(h, k, r, x, y):
    
    distancia = math.sqrt((x - h) ** 2 + (y - k) ** 2)
    
    return distancia < r

try:
    h = float(input("Introduce la coordenada x del centro del círculo amigo: "))
    k = float(input("Introduce la coordenada y del centro del círculo amigo: "))
    r = float(input("Introduce el radio del círculo :): "))
    x = float(input("Introduce la coordenada x del punto mi querido amigo: "))
    y = float(input("Introduce la coordenada y del punto mi querido amigo: "))
    
    if esta_en_el_circulo(h, k, r, x, y):
        print(f"El punto ({x}, {y}) está dentro del círculo.")
    else:
        print(f"El punto ({x}, {y}) no está dentro del círculo.")
except ValueError:
    print("Por favor, ingresa valores numéricos válidos.")
```
